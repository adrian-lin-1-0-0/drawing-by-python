```python
import matplotlib.pyplot as plt 
from matplotlib.animation import FuncAnimation
import random

class Animate_helper:
    def __init__(self,n):
        self.x = 0
        self.y = 0
        self.x_data = []
        self.y_data = []
        n_sqrt = 2*int(n ** 0.5) 
        fig, ax = plt.subplots()
        ax.set_xlim(-n_sqrt,n_sqrt)
        ax.set_ylim(-n_sqrt,n_sqrt)
        line, = ax.plot(0,0)
        self.fig = fig
        self.line = line
    
    def set_point(self):
        self.line.set_xdata(self.x_data)
        self.line.set_ydata(self.y_data)
        return self

    def walk(self):
        self.x += random.randint(-1,1)
        self.y += random.randint(-1,1)
        self.x_data.append(self.x)
        self.y_data.append(self.y)
        return self

    def get_line(self):
        return self.line,

if __name__ == "__main__":
    n = 500

    animate_helper = Animate_helper(n)
    animate = lambda _: animate_helper.walk().set_point().get_line()
    
    anim = FuncAnimation(animate_helper.fig, animate, frames=n, interval=20)
    anim.save('random-walk.gif')
    plt.show()



```
